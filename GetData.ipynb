```
import bs4
import requests as rq
r = rq.get("https://www.flipkart.com/search?q=macbook")
r.status_code 
r.text
soup=bs4.BeautifulSoup(r.text, "html.parser")
soup.findAll('div', attrs=('class','_3wU53n'))
soup.tilte
soup.title.get_text()
soup.findAll('div', attrs=('class','_3wU53n'))
Mydivs = soup.findAll('div', attrs=('class','_3wU53n'))
for item in Mydivs:
  print(item.get_text(),end="\n\n")
Mydiv = soup.findAll('div', attrs=('class','_1vC4OE _2rQ-NK'))
for item in Mydiv:
  print(item.get_text(),end="\n\n")
Mydi = soup.findAll('div', attrs=('class','_3ULzGw'))
for item in Mydi:
  print(item.get_text(),end="\n\n")
```
